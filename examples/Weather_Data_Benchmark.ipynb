```python
import pandas as pd
from ORBIT import ProjectManager, load_config
from pathlib import Path
from itertools import product

import pandas as pd
import matplotlib.pyplot as plt

from whale import Project
from whale.utilities import load_yaml
```

```python
library_path = Path("library").resolve()

weather_list = ["NOW-23_137m_lat41_03325_long-70_61667_2000to2020_time_series.csv", "WTK_137m_lat41_03325_long-70_61667_2007to2013_time_series.csv", "boem_era5_41.0_-70.5.csv"]
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
f, ax1 = plt.subplots()
colors = ["red", "forestgreen", "gold"]
count = 0
avg = list()
for i in weather_list:
    print(i)
    pd_weather_raw = pd.read_csv(library_path / "weather" / i , parse_dates=['datetime'])

    weather_data = pd_weather_raw[["datetime","windspeed_137m", "wind_direction"]]

    weather_data = weather_data[weather_data.datetime.dt.year > 2006]
    weather_data = weather_data[weather_data.datetime.dt.year < 2014]
    #display(weather_data)
    avg.append(weather_data.windspeed_137m.mean())
    
    weather_data["month"] = weather_data.datetime.dt.month



    ax1.plot(weather_data.groupby("month")["windspeed_137m"].mean(), marker = "o", color = colors[count], markeredgecolor='black')
    #ax2 = ax1.twinx()
    #ax2.plot(weather_data.groupby("month")["waveheight"].mean(), marker = "+", color = "forestgreen")
    tick_spacing = 1
    count = count + 1
ax1.grid()
ax1.set_title("Avg. wind speed by month at Vineyard Wind site (data from 2007-2013)", fontsize = 10) 
ax1.set_xlabel("Months") 
ax1.set_ylabel("Wind Speed (m/s)") 
avg= [ '%.2f' % elem for elem in avg ]
#ax2.set_ylabel("Wave Height (m)")
ax1.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing)) 
ax1.legend(["NOW-23 2007-2013 (avg = " + str(avg[0]) + " m/s)", "WTK 2007-2013 (avg = " + str(avg[1]) + " m/s)", "ERA5 2007-2013 (avg = " + str(avg[2]) + " m/s)"], loc = "lower left")
#ax2.legend(["Average wave height"], loc = "lower right")
#ax2.set_ylim(0,3.5)
ax1.set_ylim(5,12)

plt.savefig("library/plots/wind_2007_2013_weather_plot.svg")
```
