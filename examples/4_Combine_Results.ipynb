# Working Analysis For All of the Single Phase Projects

## Imports and Setup

```python
from pathlib import Path

import pandas as pd

pd.options.display.float_format = '{:,.4f}'.format
pd.options.display.max_columns = 100
pd.options.display.max_rows = 100
```

```python
library_path = Path("library").resolve()
project_order = [
    "Revolution Wind",
    "CVOW C",
    "Sunrise Wind",
    "Kitty Hawk North",
    "Marwin",
    "Momentum Wind",
    "Future Development",
    "US Wind",
    "Park City",
    "Commonwealth",
    "New England",
    "Empire Wind 1",
    "Empire Wind 2",
    "Empire Wind",
    "Atlantic Shores South 1",
    "Atlantic Shores South 2",
    "Atlantic Shores South",
    "Mayflower Wind 1",
    "Mayflower Wind 2",
    "Mayflower Wind",
    "Ocean Wind 1",
]
```

```python
outputs = pd.concat(
    [
        pd.read_csv(library_path / "results" / f"{project.replace(' ', '_')}_metrics.csv", index_col="Metrics")
        for project in project_order
    ],
    axis=1
)[project_order]
inputs = pd.concat(
    [
        pd.read_csv(library_path / "results" / f"{project.replace(' ', '_')}_inputs.csv", index_col="Inputs")
        for project in project_order
    ],
    axis=1
)[project_order]
```

```python
outputs.to_csv(library_path / "results" / "outputs.csv")
inputs.to_csv(library_path / "results" / "inputs.csv")
```
